```python
df['locality'] = df['locality'].str.title()
```

```python

lambda state: (
    "Victoria" if state == 'VIC' else
    "New South Wales" if state == 'NSW' else
    "Queensland" if state == 'QLD' else
    'South Australia' if state == 'SA' else
    'Western Australia' if state == 'WA' else
    'Australian Capital Territory' if state == 'ACT' else
    'Northern Territory' if state == 'NT' else
    'Tasmania' if state == 'TAS' else
    'NaN'
)

# df['state'] = df
```

```python
df['state'] = df['state'].apply(lambda state: (
    "Victoria" if state == 'VIC' else
    "New South Wales" if state == 'NSW' else
    "Queensland" if state == 'QLD' else
    'South Australia' if state == 'SA' else
    'Western Australia' if state == 'WA' else
    'Australian Capital Territory' if state == 'ACT' else
    'Northern Territory' if state == 'NT' else
    'Tasmania' if state == 'TAS' else
    'NaN'
))
```
